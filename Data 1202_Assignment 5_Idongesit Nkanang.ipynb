Import required libraries

```python
import glob
import pandas 
import matplotlib.pyplot 
import numpy as np
```

Import given datasets using glob function

```python
dataset = sorted(glob.glob('data*.csv'))
print(dataset)
```

Create function: visualize( ) function will plot the average, maximum, and minimum of the given dataset

```python
def visualize(filename):
    """This function identifies will plot the average, maximum, and minimum of the given dataset"""

    data = np.loadtxt(fname=filename, delimiter=',')

    fig = matplotlib.pyplot.figure(figsize=(10.5, 3.5))

    axes1 = fig.add_subplot(1, 3, 1)
    axes2 = fig.add_subplot(1, 3, 2)
    axes3 = fig.add_subplot(1, 3, 3)

    axes1.set_ylabel('average')
    axes1.plot(np.mean(data, axis=0))

    axes2.set_ylabel('max')
    axes2.plot(np.max(data, axis=0))

    axes3.set_ylabel('min')
    axes3.plot(np.min(data, axis=0))

    fig.tight_layout()
    matplotlib.pyplot.show()
```

Create function: identify_issues() The identifies_issues() function is to identify issues in the given dataset

```python
def identify_issues(filename):

    data = np.loadtxt(fname=filename, delimiter=',')

    if np.max(data, axis=0)[0] == 0 and np.max(data, axis=0)[20] == 20:
        print('I doubt the integrity of this dataset')
    elif np.sum(np.min(data, axis=0)) == 0:
        print('Why is this happening?')
    else:
        print('Dataset looks good')
```

Using 'for' loop for all the datasets and apply the visualize() and identify_issues() function to each datasets for analysis

```python
filenames = sorted(glob.glob('data*.csv'))

for filename in filenames[:4]:
    print("\n")
    print(filename)
    visualize(filename)
    identify_issues(filename)
```
