Create function: visualize( ) 
Visualize() function will plot the average, maximum, and minimum of the given dataset

```python
# function to visualize features in the given datasets

def visualize(filename):
    
    # use numpy method genfromtxt to  get the dataset into a numpy array
    file = np.genfromtxt(filename, delimiter = ",")
    x_axis = np.arange(1,41)
    # figure size
    plt.figure(figsize = (15, 30))
    fig, (x1, x2, x3) =  plt.subplots(1,3)
    # set title for the plots in subplot
    fig.suptitle(filename)
    
    # plot values in the dataset
    x1.plot(x_axis, file.mean(axis = 0)) # average values
    x2.plot(x_axis, file.max(axis = 0)) # max values
    x3.plot(x_axis, file.min(axis = 0)) # min values
    
    # provide xlab and ylab for plots
    labels = ['Feature #', 'Average', 'Max', 'Min']
    x1.set(xlabel = labels[0], ylabel = labels[1])
    x2.set(xlabel = labels[0], ylabel = labels[2])
    x3.set(xlabel = labels[0], ylabel = labels[3])

    # format and display the subplots
    fig.tight_layout()
    fig.set_figwidth(12)
    plt.show()
```

Using 'for' loop for all the datasets and apply the visualize() and identify_issues() function to each datasets for analysis

```python
for x in dataset:
    print(x)
    visualize(x)
    identify_issues(x)
```
